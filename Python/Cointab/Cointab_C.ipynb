##### Joining df order_df with sku

```python
df1 = pd.merge(order_df,sku_master_df,how= 'left', on= 'SKU')
```

```python
df1.info()
```

```python
df1.rename(columns= {'ExternOrderNo':'Order ID'}, inplace= True)
df1
```

```python
df1['Order Qty'].unique()
```

```python
# Total weight of units in a single order. 

def calc_weight(order_qty,weight):
    return order_qty * weight
```

```python
df1['total_weight'] = np.vectorize(calc_weight)(df1['Order Qty'],df1['Weight (g)'])
```

```python
df1.sample(5)
```

```python
df1.drop(columns= ['SKU','Order Qty','Weight (g)'], inplace= True)
```

```python
df1 = df1.groupby('Order ID').sum()
df1.reset_index(inplace= True)
df1
```

```python
df1.describe()
```

```python
# Converting to kg form grams

df1['total_weight'] = df1['total_weight'].apply(lambda num: round(num/1000,ndigits=2))
```

```python
# Calculating weight slabs

def weight_slab(total_weight):
    if total_weight <= 0.50:
        return 0.5
    elif total_weight <= 1:
        return 1
    elif total_weight <= 1.5:
        return 1.5
    elif total_weight <= 2:
        return 2
    elif total_weight <= 2.5:
        return 2.5
    elif total_weight <= 3:
        return 3
    elif total_weight <= 3.5:
        return 3.5
    elif total_weight <= 4:
        return 4
    else:
        return 4.5
```

```python
df1['weight_slab'] = df1['total_weight'].apply(weight_slab)
df1.head()
```

```python
df2 = pd.merge(invoice_df, pin_df, how='left', on= 'Customer Pincode')
```

```python
df2.drop(columns='Warehouse Pincode_y', inplace= True)
```

```python
df2
```

```python
def calc_diff(x,y):
    if x == y:
        return 0
    else:
        return 1

# 0 indicates no difference
```

```python
df2['zone_difference'] = np.vectorize(calc_diff)(df2['Zone_x'],df2['Zone_y'])
```

```python
df2['zone_difference'].value_counts() # 65 orders have been alloted wrong zones in invoice data
```

```python
df3 = pd.merge(df2,df1, how='left', on= 'Order ID')
df3.info()
```

```python
def fixed_charges(shipment_type,zone):
    if shipment_type == 'Forward and RTO charges':
        if zone == 'b':
            return 33 + 20.5
        elif zone == 'd':
            return 45.4 + 41.3
        else:
            return 56.6 + 50.7
    else:
        if zone == 'b':
            return 33
        elif zone == 'd':
            return 45.4
        else:
            return 56.6
```

```python
def fwd_add_charges(zone):
    if zone == 'b':
        return 28.3
    elif zone == 'd':
        return 44.8
    else:
        return 55.5
```

```python
def rto_add_charges(shipment_type,zone):
    if shipment_type == 'Forward and RTO charges':
        if zone == 'b':
            return 28.3
        elif zone == 'd':
            return 44.8
        else:
            return 55.5
    else:
        return 0
```

```python
def create_mul(slab):
    return (slab - 0.5)/0.5
```

```python
# calculating fixed charge (fwd_fixed + rto_fixed)
df3['fixed_charges'] = np.vectorize(fixed_charges)(df3['Type of Shipment'],df3['Zone_y'])

# foward additional charges
df3['fwd_add_charge'] = df3['Zone_y'].apply(fwd_add_charges)

# rto additional charges
df3['rto_add_charges'] = np.vectorize(rto_add_charges)(df3['Type of Shipment'],df3['Zone_y'])

# creating weight slab for calculation
df3['add_slab'] = df3['weight_slab'].apply(create_mul)
```

```python
# Calcuting expected charge by X

df3['calc_bill'] = df3['fixed_charges'] + (df3['add_slab'] * df3['fwd_add_charge']) + (df3['add_slab'] * df3['rto_add_charges'])
```

```python
# Calcuting Weight slab charged by Courier Company (KG)

df3['Weight slab charged by Courier Company (KG)'] = df3['Charged Weight'].apply(weight_slab)
```

```python
df3
```

```python
df3['Difference Between Expected Charges and Billed Charges (Rs.)'] = abs(round(
    df3['calc_bill'] - df3['Billing Amount (Rs.)'],2))
```

```python
df3.rename(columns={'total_weight':'Total weight as per X (KG)',
                           'weight_slab':'Weight slab as per X (KG)',
                           'Charged Weight':'Total weight as per Courier Company (KG)',
                           'Zone_y':'Delivery Zone as per X',
                           'Zone_x':'Delivery Zone charged by Courier Company',
                           'calc_bill':'Expected Charge as per X (Rs.)',
                           'Billing Amount (Rs.)':'Charges Billed by Courier Company (Rs.)'}, inplace= True)
```

```python
# Function for calculating total irregularities

def bill_accuracy(courier_bill,x_bill):
    if abs(round(courier_bill,2)) == abs(round(x_bill,2)):
        return 'Correctly Charged'
    elif abs(round(courier_bill,2)) > abs(round(x_bill,2)):
        return 'Over Charged'
    else:
        return 'Under Charged'
```

```python
df3['billing_accuracy'] = np.vectorize(bill_accuracy)(df3['Charges Billed by Courier Company (Rs.)'],
                                                   df3['Expected Charge as per X (Rs.)'])
```

```python
df3['billing_accuracy'].value_counts()
```

```python
over_chrg, correctly_chrg, under_chrg = df3['billing_accuracy'].value_counts()
```

```python
# calculating total irregularities amount

amt_correctly_chrg = round(df3[df3['billing_accuracy']
                               == 'Correctly Charged']['Charges Billed by Courier Company (Rs.)'].sum(),2)

amt_over_chrg = df3[df3['billing_accuracy']
                    == 'Over Charged']['Charges Billed by Courier Company (Rs.)'].sum() - df3[df3['billing_accuracy']
                                                                                              == 'Over Charged']['Expected Charge as per X (Rs.)'].sum()

amt_under_chrg = df3[df3['billing_accuracy']
                     == 'Under Charged']['Expected Charge as per X (Rs.)'].sum()- df3[df3['billing_accuracy']
                                                                                      == 'Under Charged']['Charges Billed by Courier Company (Rs.)'].sum()
```

```python
# Creating Output 2 data.

data = {'Count': [correctly_chrg, over_chrg, under_chrg],
        'Amount (Rs)': [amt_correctly_chrg, amt_over_chrg, amt_under_chrg]}
  
# Creates pandas DataFrame.
output_2_df = pd.DataFrame(data, index=['Total orders where X has been correctly charged',
                               'Total Orders where X has been overcharged',
                               'Total Orders where X has been undercharged'])
```
